#**Statistics Advance Part 2 Theory Questions**

**1. What is hypothesis testing in statistics?**

Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves formulating two competing hypotheses and using sample data to determine which is more likely.

**2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**

* **Null Hypothesis (H₀):** Assumes no effect or no difference; it's the default or status quo.
* **Alternative Hypothesis (H₁ or Ha):** Represents the claim you want to test; indicates a significant effect or difference.

For example:

* H₀: μ = 50
* H₁: μ ≠ 50 (two-tailed) or μ > 50 / μ < 50 (one-tailed)

**3. What is the significance level in hypothesis testing, and why is it important?**

The **significance level (α)** is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values are 0.05, 0.01, and 0.10. It sets the threshold for deciding whether the observed result is statistically significant.

**4. What does a P-value represent in hypothesis testing?**

The **P-value** is the probability of obtaining results as extreme as the observed results, assuming the null hypothesis is true. A small P-value indicates strong evidence against the null hypothesis.

**5. How do you interpret the P-value in hypothesis testing?**

* If **P ≤ α**: Reject the null hypothesis (significant result)
* If **P > α**: Fail to reject the null hypothesis (not significant)

Example: P = 0.03, α = 0.05 → Reject H₀

**6. What are Type 1 and Type 2 errors in hypothesis testing?**

* **Type I Error (False Positive):** Rejecting H₀ when it is actually true (controlled by α).
* **Type II Error (False Negative):** Failing to reject H₀ when it is actually false (denoted by β).

**7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**

* **One-tailed test:** Tests for deviation in one direction (e.g., H₁: μ > μ₀).
* **Two-tailed test:** Tests for deviation in both directions (e.g., H₁: μ ≠ μ₀).

**8. What is the Z-test, and when is it used in hypothesis testing?**

The **Z-test** is used when:

* The population standard deviation is known.
* The sample size is large (n > 30).
  It's used to compare means, proportions, or to test a single mean against a population mean.

**9. How do you calculate the Z-score, and what does it represent in hypothesis testing?**

**Z-score formula:**

$$
Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
$$

Where:

* $\bar{x}$ = sample mean
* $\mu$ = population mean
* $\sigma$ = population standard deviation
* $n$ = sample size

It tells how many standard deviations a sample mean is from the population mean.

**10. What is the T-distribution, and when should it be used instead of the normal distribution?**

The **T-distribution** is used when:

* The population standard deviation is unknown.
* The sample size is small (n < 30).

It has heavier tails than the normal distribution, accounting for more variability in small samples.

**11. What is the difference between a Z-test and a T-test?**

The **main difference** between a Z-test and a T-test lies in the knowledge of population standard deviation and the sample size:

* Use a **Z-test** when:

  * The **population standard deviation** is known.
  * The **sample size is large** (typically n > 30).
  * The data is approximately normally distributed.

* Use a **T-test** when:

  * The **population standard deviation is unknown**.
  * The **sample size is small** (typically n < 30).
  * The data comes from a population that is approximately normal.

In essence, the T-test is a more flexible version of the Z-test used when there is more uncertainty (due to smaller sample size or unknown variability).

**12. What is the T-test, and how is it used in hypothesis testing?**

A **T-test** is used to compare means when population standard deviation is unknown. Types include:

* One-sample t-test
* Two-sample (independent) t-test
* Paired sample t-test

**13. What is the relationship between Z-test and T-test in hypothesis testing?**

Both tests compare sample means to population means but differ in assumptions:

* T-test is a generalization of Z-test used for small samples or unknown variance.
* As sample size increases, T-distribution approximates the normal (Z) distribution.

**14. What is a confidence interval, and how is it used to interpret statistical results?**

A **confidence interval (CI)** provides a range of plausible values for a population parameter. A 95% CI means that if the experiment is repeated many times, 95% of the CIs would contain the true parameter.

**15. What is the margin of error, and how does it affect the confidence interval?**

The **margin of error (MOE)** is the amount added/subtracted from the point estimate to create the CI:

$$
\text{CI} = \text{estimate} \pm \text{margin of error}
$$

Larger MOE = wider interval = less precision.

**16. How is Bayes' Theorem used in statistics, and what is its significance?**

**Bayes' Theorem** updates the probability of a hypothesis given new evidence:

$$
P(H|E) = \frac{P(E|H) \cdot P(H)}{P(E)}
$$

Used in fields like medical testing, spam filtering, and machine learning.

**17. What is the Chi-square distribution, and when is it used?**

The **Chi-square distribution** is used in:

* Testing independence between categorical variables.
* Testing goodness of fit of a model.
  It is skewed right and defined only for non-negative values.

**18. What is the Chi-square goodness of fit test, and how is it applied?**

This test checks if observed categorical data matches expected distribution:

$$
\chi^2 = \sum \frac{(O - E)^2}{E}
$$

Where $O$ = observed, $E$ = expected. If the value is too large, reject H₀.

**19. What is the F-distribution, and when is it used in hypothesis testing?**

The **F-distribution** is used to compare two variances and is the basis of ANOVA. It's right-skewed and depends on two degrees of freedom (numerator and denominator).

**20. What is an ANOVA test, and what are its assumptions?**

**ANOVA (Analysis of Variance)** tests if there are significant differences among group means.
Assumptions:

* Independence of observations
* Normality of residuals
* Homogeneity of variances (equal variances)

**21. What are the different types of ANOVA tests?**

* **One-way ANOVA:** One independent variable
* **Two-way ANOVA:** Two independent variables
* **Repeated Measures ANOVA:** Same subjects tested multiple times.

**22. What is the F-test, and how does it relate to hypothesis testing?**

An **F-test** compares the variances of two populations or models. It is used in:

* ANOVA (to test mean differences via variance)
* Regression analysis (to test overall model fit)

```python
#3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

from scipy.stats import bernoulli

def bernoulli_pdf(p):
    """
    Calculate the PDF (PMF) of a Bernoulli distribution for p.
    Returns probabilities for outcomes 0 and 1.
    """
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1")

    prob_0 = bernoulli.pmf(0, p)
    prob_1 = bernoulli.pmf(1, p)

    return {0: prob_0, 1: prob_1}

# Example usage
pdf = bernoulli_pdf(0.6)
print("Bernoulli PDF:", pdf)
```

```python
#14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    z = (x - mean) / std_dev
    return z

# Example data
x = 75             # Data point
mean = 70          # Population mean
std_dev = 10       # Population standard deviation

# Calculate Z-score
z_score = calculate_z_score(x, mean, std_dev)
print(f"Z-score for x = {x}: {z_score:.2f}")

# Plot standard normal distribution
x_values = np.linspace(-4, 4, 1000)
y_values = norm.pdf(x_values, 0, 1)

plt.plot(x_values, y_values, label="Standard Normal PDF", color="blue")

# Mark the Z-score on the curve
plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')
plt.fill_between(x_values, 0, y_values, where=(x_values <= z_score), color='red', alpha=0.3)

plt.title("Z-score on Standard Normal Distribution")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#15.Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

# Sample data
sample = [72, 68, 75, 70, 74, 69, 71, 73, 70, 72]

# Population parameters
population_mean = 70
population_std = 3  # Assumed known for Z-test

# Sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)

# Z-statistic calculation
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-score = {z_score:.4f}")
print(f"P-value = {p_value:.4f}")

# Significance level
alpha = 0.05

# Decision
if p_value < alpha:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

```python
#16.Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy.stats import norm

# Sample dataset
data = [72, 68, 75, 70, 74, 69, 71, 73, 70, 72]

# Sample statistics
sample_mean = np.mean(data)
sample_size = len(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation

# Confidence level
confidence = 0.95
z_critical = norm.ppf((1 + confidence) / 2)

# Margin of error
margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))

# Confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval (95%): ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution_skew_kurtosis(size=1000, mean=0, std_dev=1):
    # Generate data from a normal distribution
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate skewness and kurtosis
    skewness = skew(data)
    kurt = kurtosis(data)  # By default, this is excess kurtosis (kurtosis - 3)

    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (excess): {kurt:.4f}")

    # Interpretation
    if abs(skewness) < 0.5:
        print("Data is approximately symmetric.")
    else:
        print("Data is skewed.")

    if abs(kurt) < 0.5:
        print("Kurtosis is close to normal (mesokurtic).")
    elif kurt > 0:
        print("Data is leptokurtic (peaked).")
    else:
        print("Data is platykurtic (flat).")

# Example usage
analyze_distribution_skew_kurtosis()
```

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

# Sample data (e.g., exam scores from a class)
sample = [72, 68, 75, 70, 74, 69, 71, 73, 70, 72]

# Known population parameters
population_mean = 70
population_std = 3  # Assumed known

# Sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)

# Z-test calculation
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Output
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

# Significance level
alpha = 0.05

# Interpretation
if p_value < alpha:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population mean.")
```

```python
#3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Hypothesis test decision
    if p_value < alpha:
        print("Reject the null hypothesis: Significant difference.")
    else:
        print("Fail to reject the null hypothesis: No significant difference.")

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=102, scale=10, size=30)  # Simulated sample
population_mean = 100
population_std = 10  # Known

one_sample_z_test(sample_data, population_mean, population_std)
```

```python
#4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- Parameters ---
sample = np.random.normal(loc=102, scale=10, size=30)  # Simulated sample
population_mean = 100
population_std = 10  # Known
alpha = 0.05

# --- Z-Test ---
sample_mean = np.mean(sample)
n = len(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# --- Critical Z for two-tailed test ---
z_critical = norm.ppf(1 - alpha/2)

# --- Output results ---
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Z-critical (±): ±{z_critical:.2f}")

# --- Plot standard normal distribution ---
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Curve', color='blue')

# Shade critical regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region (Left)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.3, label='Rejection Region (Right)')

# Mark z-score
plt.axvline(z_score, color='green', linestyle='--', label=f'Z = {z_score:.2f}')

plt.title("Two-Tailed Z-Test Visualization")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()

# --- Conclusion ---
if abs(z_score) > z_critical:
    print("Reject the null hypothesis: Significant difference.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_null=100, mu_alt=105, sigma=10, n=30, alpha=0.05):
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value (two-tailed)
    z_crit = norm.ppf(1 - alpha/2)

    # Critical region bounds under H0
    crit_low = mu_null - z_crit * se
    crit_high = mu_null + z_crit * se

    # x-axis range covering both distributions
    x = np.linspace(mu_null - 4*se, mu_alt + 4*se, 1000)

    # PDF for H0 and H1
    h0_pdf = norm.pdf(x, mu_null, se)
    h1_pdf = norm.pdf(x, mu_alt, se)

    # Plot both distributions
    plt.figure(figsize=(10, 5))
    plt.plot(x, h0_pdf, label="H₀ Distribution", color='blue')
    plt.plot(x, h1_pdf, label="H₁ Distribution", color='green')

    # Type I Error region (α)
    plt.fill_between(x, h0_pdf, where=((x < crit_low) | (x > crit_high)), color='red', alpha=0.3, label='Type I Error (α)')

    # Type II Error region (β)
    plt.fill_between(x, h1_pdf, where=((x >= crit_low) & (x <= crit_high)), color='orange', alpha=0.4, label='Type II Error (β)')

    # Labels and legend
    plt.axvline(crit_low, color='black', linestyle='--', label=f"Critical Value (Low) = {crit_low:.2f}")
    plt.axvline(crit_high, color='black', linestyle='--', label=f"Critical Value (High) = {crit_high:.2f}")
    plt.title("Type I and Type II Errors Visualization")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Calculate Type II Error (β)
    beta = norm.cdf(crit_high, loc=mu_alt, scale=se) - norm.cdf(crit_low, loc=mu_alt, scale=se)
    power = 1 - beta

    # Output summary
    print(f"Critical region: ({crit_low:.2f}, {crit_high:.2f})")
    print(f"Type I Error (α): {alpha:.2f}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {power:.4f}")

# Example usage
visualize_type1_type2_errors()
```

```python
#6.Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy.stats import ttest_ind

# Simulate data for two independent groups
np.random.seed(0)
group1 = np.random.normal(loc=75, scale=10, size=30)  # Group 1 (e.g., Method A)
group2 = np.random.normal(loc=80, scale=12, size=30)  # Group 2 (e.g., Method B)

# Perform independent t-test (equal variances assumed by default)
t_stat, p_value = ttest_ind(group1, group2)

# Output results
print(f"Group 1 Mean: {np.mean(group1):.2f}")
print(f"Group 2 Mean: {np.mean(group2):.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The group means are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between group means.")
```

```python
#7.Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

# Step 1: Simulate paired data (e.g., before and after treatment)
np.random.seed(0)
before = np.random.normal(loc=70, scale=8, size=30)
after = before + np.random.normal(loc=2, scale=5, size=30)  # Small improvement

# Step 2: Perform paired T-test
t_stat, p_value = ttest_rel(before, after)

# Step 3: Output results
print(f"Mean (Before): {np.mean(before):.2f}")
print(f"Mean (After): {np.mean(after):.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Significant difference between paired means.")
else:
    print("Fail to reject the null hypothesis: No significant difference between paired means.")

# Step 4: Visualization
plt.figure(figsize=(8, 5))
x = np.arange(len(before))
plt.plot(x, before, 'o-', label='Before', color='blue')
plt.plot(x, after, 'o-', label='After', color='green')
plt.title("Paired Sample Comparison")
plt.xlabel("Subject")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#8. Simulate data and perform both Z-test and T-test, then compare the results using Python.

import numpy as np
from scipy.stats import ttest_1samp, norm

# Step 1: Simulate sample data from a normal distribution
np.random.seed(42)
sample = np.random.normal(loc=102, scale=12, size=30)

# Assume population parameters
population_mean = 100
population_std = 12  # Known std for Z-test

# Step 2: Z-test
sample_mean = np.mean(sample)
n = len(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
z_p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed

# Step 3: T-test
t_stat, t_p_value = ttest_1samp(sample, popmean=population_mean)

# Step 4: Print and compare results
print("=== Z-Test (Population Std Dev Known) ===")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value (Z-test): {z_p_value:.4f}")

print("\n=== T-Test (Population Std Dev Unknown) ===")
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value (T-test): {t_p_value:.4f}")

# Optional: Decision
alpha = 0.05
print("\n--- Hypothesis Decision at α = 0.05 ---")
if z_p_value < alpha:
    print("Z-Test: Reject the null hypothesis.")
else:
    print("Z-Test: Fail to reject the null hypothesis.")

if t_p_value < alpha:
    print("T-Test: Reject the null hypothesis.")
else:
    print("T-Test: Fail to reject the null hypothesis.")
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # Standard error of the mean

    # t critical value for given confidence level
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    # Margin of error
    margin = t_crit * std_err

    lower_bound = mean - margin
    upper_bound = mean + margin

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    return (lower_bound, upper_bound)

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=15, size=30)
confidence_interval(sample_data)
```

```python
#10.Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    standard_error = sample_std / np.sqrt(n)

    # t critical value for the confidence level
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    # Margin of Error
    margin_of_error = t_crit * standard_error

    print(f"Sample Size (n): {n}")
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Sample Standard Deviation: {sample_std:.2f}")
    print(f"Standard Error: {standard_error:.2f}")
    print(f"Margin of Error at {int(confidence * 100)}% confidence: ±{margin_of_error:.2f}")

    return margin_of_error

# Example usage
np.random.seed(1)
sample_data = np.random.normal(loc=50, scale=10, size=25)
calculate_margin_of_error(sample_data)
```

```python
#11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior, sensitivity, false_positive):
    """Calculate P(Disease | Positive) using Bayes' Theorem."""
    # Complementary values
    not_prior = 1 - prior
    specificity = 1 - false_positive

    # Marginal probability of testing positive
    p_positive = (sensitivity * prior) + (false_positive * not_prior)

    # Posterior probability
    posterior = (sensitivity * prior) / p_positive

    print(f"Prior P(Disease): {prior}")
    print(f"Sensitivity P(Positive | Disease): {sensitivity}")
    print(f"False Positive Rate P(Positive | No Disease): {false_positive}")
    print(f"Posterior P(Disease | Positive): {posterior:.4f}")

    return posterior

# Example usage
bayes_theorem(prior=0.01, sensitivity=0.99, false_positive=0.05)
```

Explanation of the Process:

* Prior belief: There's a 1% chance a person has the disease.

* Data (test result): The person tests positive.

* Likelihood: The test is 99% accurate for those who have the disease.

* False positive: 5% of healthy people test positive.

* Posterior: After seeing the test result, the probability of having the disease rises to ~16.7% — not 99%, which surprises many!

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python.

import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table
# Example: Gender vs Preference
data = pd.DataFrame({
    'Tea': [30, 10],
    'Coffee': [20, 40]
}, index=['Male', 'Female'])

print("Contingency Table:")
print(data)

# Step 2: Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

# Step 3: Output results
print("\nChi-Square Test Results:")
print(f"Chi2 Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=data.index, columns=data.columns))

# Step 4: Interpretation
alpha = 0.05
if p < alpha:
    print("\nConclusion: Reject the null hypothesis — variables are dependent.")
else:
    print("\nConclusion: Fail to reject the null hypothesis — variables are independent.")
```

```python
#13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import pandas as pd
import numpy as np

# Step 1: Define the observed data (contingency table)
# Example: Product preference by region
observed = pd.DataFrame({
    'Product A': [20, 30, 25],
    'Product B': [25, 15, 35]
}, index=['Region 1', 'Region 2', 'Region 3'])

print("Observed Frequencies:")
print(observed)

# Step 2: Calculate row totals, column totals, and grand total
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
grand_total = observed.values.sum()

# Step 3: Calculate expected frequencies
expected = pd.DataFrame(index=observed.index, columns=observed.columns)

for row in observed.index:
    for col in observed.columns:
        expected.loc[row, col] = (row_totals[row] * col_totals[col]) / grand_total

expected = expected.astype(float)

# Step 4: Display results
print("\nExpected Frequencies:")
print(expected.round(2))
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
from scipy.stats import chisquare

# Step 1: Observed frequencies (e.g., outcomes of rolling a die)
observed = np.array([18, 20, 17, 22, 19, 24])  # observed counts for each face of a die

# Step 2: Expected frequencies (uniform distribution: equal chance for each face)
expected = np.array([20, 20, 20, 20, 20, 20])

# Step 3: Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Print results
print("Observed Frequencies:", observed)
print("Expected Frequencies:", expected)
print(f"Chi-Square Statistic: {chi_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 5: Interpret results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: Observed data fits the expected distribution.")
```
